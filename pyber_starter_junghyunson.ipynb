```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_df = pd.read_csv(city_data_to_load)
ride_df = pd.read_csv(ride_data_to_load)
# city_df.head() 	city	driver_count	type
# ride_df.head()    city	date	fare	ride_id
# Combine the data into a single dataset
merged_df = pd.merge(ride_df,city_df, on ="city", how="left")


# Display the data table for preview
merged_df.head()
```

```python
# Obtain the x and y coordinates for each of the three city types
# Urban
urban_x = urban_total_rides
urban_y = urban_average_fare
#Suburban
suburban_x = suburban_total_rides
suburban_y = suburban_average_fare
#Rural
rural_x = rural_total_rides
rural_y = rural_average_fare


# Build the scatter plots for each city types
urban_plt = plt.scatter(urban_x, urban_y, marker="o", facecolors="coral", 
                        edgecolors="black",s=urban_driver_count*10, alpha=0.75, label="Urban",linewidths=1)
suburban_plt = plt.scatter(suburban_x, suburban_y, marker="o", facecolors="skyblue", 
                        edgecolors="black",s=suburban_driver_count*10, alpha=0.75, label="Suburban",linewidths=1)
rural_plt = plt.scatter(rural_x, rural_y, marker="o", facecolors="gold", 
                        edgecolors="black",s=rural_driver_count*10, alpha=0.75, label="Rual",linewidths=1)

# Incorporate the other graph properties

plt.grid()
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")

# Create a legend
legend = plt.legend(title="City Types", fontsize=8)

legend.legendHandles[0]._sizes = [30]
legend.legendHandles[1]._sizes = [30]
legend.legendHandles[2]._sizes = [30]
    
# Incorporate a text label regarding circle size
plt.text(42,36.5,"Note:")
plt.text(42,35,"Circle size correlates with driver count per city.")

# Save Figure

plt.savefig("Pyber_Ride_Sharing_Data.png")

plt.show()
```

## Total Fares by City Type

```python
# help(plt.pie)
```

```python
# Calculate Type Percents
urban_total_fare = merged_df[merged_df["type"]=="Urban"].fare.sum()
suburban_total_fare = merged_df[merged_df["type"]=="Suburban"].fare.sum()
rural_total_fare = merged_df[merged_df["type"]=="Rural"].fare.sum()
total_fare = merged_df["fare"].sum()

urban_total_fare_percentage = urban_total_fare/total_fare*100
suburban_total_fare_percentage = suburban_total_fare/total_fare*100
rural_total_fare_percentage = rural_total_fare/total_fare*100

# Build Pie Chart

x=[urban_total_fare_percentage,rural_total_fare_percentage,suburban_total_fare_percentage]
labels = ["Urban","Rural","Suburban"]
explode=[0.1,0,0]
colors = ["coral","gold","skyblue"]
plt.pie(x,labels=labels, explode=explode, shadow=True, startangle=290,autopct="%1.1f%%", colors=colors)
plt.title("% of Total Fares by City Type")
# Save Figure

plt.savefig("Percentage_of_Total_Fares_by_City_Type.png")
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
urban_total_rides = merged_df[merged_df["type"]=="Urban"].ride_id.count()
# urban_total_rides
suburban_total_rides = merged_df[merged_df["type"]=="Suburban"].ride_id.count()
rural_total_rides = merged_df[merged_df["type"]=="Rural"].ride_id.count()
city_total_rides = merged_df["ride_id"].count()
# city_total_rides

urban_total_rides_percentage = urban_total_rides/city_total_rides*100
suburban_total_rides_percentage = suburban_total_rides/city_total_rides*100
rural_total_rides_percentage = rural_total_rides/city_total_rides*100

# Build Pie Chart
x=[urban_total_rides_percentage,rural_total_rides_percentage,suburban_total_rides_percentage]
labels = ["Urban","Rural","Suburban"]
explode=[0.1,0,0]
colors = ["coral","gold","skyblue"]
plt.pie(x,labels=labels, explode=explode, shadow=True, startangle=260,autopct="%1.1f%%", colors=colors)
plt.title("% of Total Rides by City Type")

# Save Figure
plt.savefig("Percentage_of_Total_Rides_by_City_Type.png")
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
merged_df_no_dups = merged_df.drop_duplicates("city")
urban_total_driver = merged_df_no_dups[merged_df_no_dups["type"]=="Urban"].driver_count.sum()
suburban_total_driver = merged_df_no_dups[merged_df_no_dups["type"]=="Suburban"].driver_count.sum()
rural_total_driver = merged_df_no_dups[merged_df_no_dups["type"]=="Rural"].driver_count.sum()
city_total_driver = merged_df_no_dups["driver_count"].sum()

# print(urban_total_driver)
# print(suburban_total_driver)
# print(rural_total_driver)
# print(city_total_driver)
# city_total_rides

urban_total_driver_percentage = urban_total_driver/city_total_driver*100
suburban_total_driver_percentage = suburban_total_driver/city_total_driver*100
rural_total_driver_percentage = rural_total_driver/city_total_driver*100

# print(urban_total_driver_percentage)
# print(suburban_total_driver_percentage)
# print(rural_total_driver_percentage)

# Build Pie Charts
x=[urban_total_driver_percentage,rural_total_driver_percentage,suburban_total_driver_percentage]
labels = ["Urban","Rural","Suburban"]
explode=[0.1,0,0]
colors = ["coral","gold","skyblue"]
plt.pie(x,labels=labels, explode=explode, shadow=True, startangle=230,autopct="%1.1f%%", colors=colors)
plt.title("% of Total Drivers by City Type")

# Save Figure
plt.savefig("Percentage_of_Total_Drivers_by_City_Type.png")
plt.show()
```

## Observation

1. While it's not clear why the average fare is higher in the rural area (it may be due to a longer drive distance or the lack of drivers), the bubble chart suggests that it is highest in the rural area than the urban and the suburban areas. 
2. The total number of drivers is highest in the urban area due to higher demand and it makes up 62.7% of the total fare. This suggests that the urban market may be saturated but it will require further research to confirm it. 
3. The drivers in the suburban area make up 16.5% of the total drivers and earn 30.5% of the total fare. This suggests that the profit margin in this area is higher than the other two and it can be a good market for us to target.
